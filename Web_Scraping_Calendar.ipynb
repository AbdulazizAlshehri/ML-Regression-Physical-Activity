# Scrapping Georgian to Islamic Lunar (Hijry)  Calendar

Unlike Georgian calendar that have fixed characteristics, Islamic (Hijry) calendar is lunar calendar. Lunar calendar is the official calendar in Saudi Arabia education system. All school vacations, religous events are scheduled according to Lunar Calendar.

In this file we scrapped a mapping table (Georgian day and equivalent lunar date) for 40 years.
The result will be a conversion table + marking school vacation days and religous events


```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
from datetime import datetime
import time


Hijri = []
Meladi =[]
cols = ['HijryDate','MeladiDate']


#loop list of months 
Hijry_Months = ['Rabiulawal','Safar','Muharam','Zulhijah','Zulkaedah','Syawal','Ramadan','Syaaban','Rejab','Jamadilakhir','Jamadilawal','Rabiulakhir']
#loop years , from 1420 (1999) to 1460 (2040) 
for yr in range(1420,1461):
    for mnth in Hijry_Months:
        #url = "https://hijri-calendar.com/Rabiulakhir/1442"
        url = "https://hijri-calendar.com/"+mnth+"/"+str(yr)
        response = requests.get(url)
        soup = BeautifulSoup(response.text, "lxml")
        print(url)
        for  el in soup.find_all(class_='hijri'):
            Hijri.append(el.text)

        skip = 0 
        el_list = soup.find_all(class_='meladi')
        for el in range(1,len(el_list)):
            #weekday, name of event, and meladi date all have meladi class
            try:
                #Add date only if we can parse it.
                datetime.strptime(el_list[el].text.strip(), "%Y/%m/%d")
                Meladi.append(el_list[el].text.strip())
            except:
                #print(el_list[el].text.strip())
                continue
        #check size every time to stop iteration in case of error
        assert len(Meladi) == len(Hijri)
        time.sleep(1)

        #print(el_list[el])
```

```python
# Create data frame from two lists

df = pd.DataFrame(columns=cols,data=list(zip(Hijri,Meladi)))
df
```

```python
# Parse, extract year, month, day, weekday
df["Date"] = pd.to_datetime(df['MeladiDate'])
df["AD_Year"] = df["Date"].dt.year
df["AD_Month"] = df["Date"].dt.month
df["AD_Day"] = df["Date"].dt.day
df["AD_WeekdayNum"] = df['Date'].dt.dayofweek
# split Lunar calendar into its components
df[['AH_Year','AH_Month','AH_Day']] = (df['HijryDate'].str.split('/',expand=True)).astype('int')

df
```

```python
#remove duplicates 
df.drop_duplicates('Date',inplace=True)

df
```
