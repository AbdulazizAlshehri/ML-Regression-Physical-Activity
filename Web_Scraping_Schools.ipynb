# Scrapping school vacations calendar 

This file contain the code to scrap published calendar from  Ministry of Education website using Selenium.

Unfortunatelly, we could not use scrapped data programmatically since MoE was not consistent in written descriptions

```python
from bs4 import BeautifulSoup
import requests
import time, os
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
chromedriver = "c:\\bin\\chromedriver.exe" # path to the chromedriver executable
os.environ["webdriver.chrome.driver"] = chromedriver

# the official Ministry of education site uses dynamic content. hence, selenium will be used to simulate mouse clicks
url = "https://moe.gov.sa/ar/education/generaleducation/Pages/academicCalendar.aspx"


driver = webdriver.Chrome(chromedriver)
driver.get(url)

soup = BeautifulSoup(driver.page_source, 'html.parser')
```

```python
# first find years id to click it 
years_ids = []
for el in soup.find_all(class_='ACYearsItem'):
    years_ids.append([el['id'],el['year']])
years_ids
```

```python
# loop through table and add to array 
cols = ['Desc','Hijri','Meladi']
rows= []

for yr in years_ids:
    cls = "ACYearsItem-"+str(yr[0])
    print(cls)
    filter_button = driver.find_element_by_xpath(f"//span[contains(@class,'{cls}')]")
    filter_button.click()
    #wait 5 seconds for ajax to load
    time.sleep(5)
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    
    for el in soup.find(class_='table-striped').tbody.find_all("tr"):
        td_list = el.select('td')
        rows.append([td_list[0].text,td_list[1].text,td_list[2].text])

        #print(el.td.text)
    df = pd.DataFrame(columns=cols,data=rows)
df
```

```python
df.to_csv("School-Calendar.csv")
```
